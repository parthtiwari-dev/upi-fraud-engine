```python
import duckdb

con = duckdb.connect('../data/processed/transactions.duckdb')

print("Total rows:", con.execute("SELECT count(*) FROM transactions").fetchone()[0])
```

```python
# Find a device_id that is part of a device_ring
target_device = con.execute("""
    SELECT device_id 
    FROM transactions 
    WHERE fraud_pattern = 'device_ring' 
    LIMIT 1
""").fetchone()[0]

print(f"Investigating Device Ring: {target_device}")

# Get all users using this device
ring_members = con.execute(f"""
    SELECT DISTINCT payer_id 
    FROM transactions 
    WHERE device_id = '{target_device}'
""").df()

print(f"Users sharing this device: {len(ring_members)}")
print(ring_members)

# VERDICT
if len(ring_members) > 1:
    print(f" SUCCESS: {len(ring_members)} different users shared device {target_device}. Device Ring Confirmed!")
else:
    print(" FAIL: Only 1 user found. Injection logic might be wrong.")
```

```python
# Find a user with velocity spike
target_user = con.execute("""
    SELECT payer_id 
    FROM transactions 
    WHERE fraud_pattern = 'velocity_spike' 
    LIMIT 1
""").fetchone()[0]

print(f"Investigating Velocity Spike for User: {target_user}")

# Get their transactions sorted by time
spike_txns = con.execute(f"""
    SELECT event_timestamp, is_fraud, fraud_pattern 
    FROM transactions 
    WHERE payer_id = '{target_user}'
    ORDER BY event_timestamp
""").df()

print(spike_txns)

# VERDICT
# Visually check if timestamps are close together (e.g., 30 seconds apart)
```

```python
times = con.execute("""
    SELECT min(event_timestamp) as start_time, max(event_timestamp) as end_time 
    FROM transactions
""").df()

print("Dataset Start:", times['start_time'][0])
print("Dataset End:  ", times['end_time'][0])
```
